```python
# CS HW3 , Clair McLafferty (cm2h)
# Kip McCharen (cm2rh)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r"http://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data", header=None)
df.columns = ['class_id','alcohol','malic_acid','ash','ash_alkalinity','magnesium','total_phenols','flavanoids','nonflavanoid_phenols','proanthocyanins','color_intensity','hue','od280_od315_diluted','proline']
print(df.head(5))
print(len(df.index))
```

# Line Plot

A density ploty of alcohol in the Italian wine dataset shows a peculiar dip in the middle but otherwise relatively normal distribution.

```python
plt.style.use('seaborn') #classic seaborn style
df['alcohol'].plot.density()
```

# Bar Plot

These bars represent the separate classes of Italian wine. 

There is a slight but noticeable difference in alcohol between them. 

```python
sns.barplot(x='alcohol', y='class_id', data=df, orient='h') #horizontal orientation on its side
```

# Histogram

The histogram below of 50 even groupings of alcohol content of Italian wines shows a relatively normal distribution, although again there is an overall dip in the center. 

```python
plt.style.use('classic')
sns.distplot(df.alcohol, bins=50, color='r') # 100 bins, and r= red color
```

## Scatterplot
This dataset consists of three unknown cultivars of Italian wine. 

The chart clearly demonstrates a distinct separation by flavanoid and alcohol content together. 

```python
plt.style.use("dark_background") #dark background to see the white dots
sns.scatterplot('alcohol', 'flavanoids', hue='class_id', palette='Oranges', data=df)  
plt.title('Changes in alcohol %s versus flavanoids across 3 different cultivars of Italian Wine')
```

# Box Plot

Below we can see the distributions of alcohol across classes more easily as box plots. 

```python
sns.catplot(x='class_id', y='alcohol', kind='box',
               data=df)
```
